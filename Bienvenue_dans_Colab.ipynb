```
import pandas as pd

contenu = pd.read_csv("/content/resultats-elections-presidentielles-2022-1er-tour.csv")
print(contenu.head())  # affiche les 5 premières lignes
```

```
nb_lignes = len(contenu)
nb_colonnes = len(contenu.columns)

print("Nombre de lignes :", nb_lignes)
print("Nombre de colonnes :", nb_colonnes)
```

```
types_colonnes = {}

for col in contenu.columns:
    t = contenu[col].dtype

    # Conversion des types NumPy → types lisibles
    if t == "int64":
        types_colonnes[col] = "int"
    elif t == "float64":
        types_colonnes[col] = "float"
    elif t == "bool":
        types_colonnes[col] = "bool"
    else:
        types_colonnes[col] = "str"  # object, string…

# Affichage propre
for colonne, type_var in types_colonnes.items():
    print(f"{colonne} : {type_var}")
```
